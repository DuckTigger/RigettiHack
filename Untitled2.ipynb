```python
import numpy as np
from numpy.random import random_integers as rand
import pyquil
from pyquil.quil import Program
from pyquil.api import QVMConnection, CompilerConnection
from pyquil.gates import *
from scipy import linalg
from grove.tomography.state_tomography import do_state_tomography
from grove.tomography.utils import notebook_mode
from pyquil.api import QPUConnection, get_devices, Job
import json
from mock import MagicMock
from __future__ import print_function
import matplotlib.pyplot as plt


acorn = get_devices(as_dict=True)['19Q-Acorn']
qpu = QPUConnection(acorn)
qvm= QVMConnection()
```

```python
qubits = [6,11, 16]
bell_state_program = Program(H(qubits[2]),
                             CNOT(qubits[2],qubits[0]),
                             H(qubits[1]),
                             CNOT(qubits[1],qubits[2]),
                             )
```

```python
def compile_INIT_gate():   
    p = bell_state_program
    compiler = CompilerConnection(acorn)
    compiledProg = compiler.compile(p)
    qpuProg = Program()
    qpuProg.inst(compiledProg)
    print(qvm.wavefunction(qpuProg))
    return qpuProg
bell_state_program = compile_INIT_gate()
```

```python
print(qvm.wavefunction(bell_state_program))

NUM_SAMPLES=2000
```

```python
state_tomography_qpu, _, _ = do_state_tomography(bell_state_program, NUM_SAMPLES, qpu, qubits)

state_tomography_qvm, _, _ = do_state_tomography(bell_state_program, NUM_SAMPLES, qvm, qubits)
```

```python
state_fidelity = state_tomography_qpu.fidelity(state_tomography_qvm.rho_est)

#if not SEND_PROGRAMS:
 #   EPS = .01
  #  assert np.isclose(state_fidelity, 1, EPS)

qpu_plot = state_tomography_qpu.plot();
qpu_plot.text(0.35, 0.9, r'$Fidelity={:1.1f}\%$'.format(state_fidelity*100), size=20)
plt.savefig("qpu_init.eps")
state_tomography_qvm.plot();
plt.savefig("qvm_init.eps")
```
